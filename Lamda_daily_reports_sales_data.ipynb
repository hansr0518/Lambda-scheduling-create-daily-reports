```python
conn = connect(s3_staging_dir = 's3://athena-results/',region_name = 'us-east-1')
```

```python
df = pd.read_sql("SELECT * FROM sales_db.sales_customer limit 100", conn)
df.head()
```

```python
query1="""SELECT current_date AS today,
            date(split_part(record_timestamp_new,' ',1))  AS date,
            customer_id,
            count(customer_id) AS cnt_customer
        FROM sales_db.sales_customer
        WHERE date(split_part(record_timestamp_new,' ',1))
                    BETWEEN current_date - interval '3' month
                    AND current_date 
        GROUP BY  date(split_part(record_timestamp_new,' ',1)) ,customer_id
        ORDER by date,customer_id"""
```

```python
df = pd.read_sql(query1, conn)
df.head()
```
